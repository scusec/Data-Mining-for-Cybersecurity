```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values # 特征矩阵
y = dataset.iloc[:, 3].values # 依赖向量

print(dataset)
print('------------------------------')
print(dataset.iloc[3])
print('------------------------------')
print(dataset.iloc[3, :2])
print('------------------------------')
print(X)
print('------------------------------')
print(y)
```

```python
from sklearn.preprocessing import Imputer
# Imputer处理丢失的数据，例如用mean（平均值）、median（中位数）或most_frequent（最常见）代替缺失值
# imputer对象参数：missing_values（缺失值的占位符，设定成整数或是NaN查找缺失值）
# strategy（差补策略，上述三种）axis（0沿列插入，1沿行插入）
imputer = Imputer(missing_values="NaN", strategy="mean", axis=0)
imputer = imputer.fit(X[:, 1:3]) # 与数据先匹配
X[:, 1:3] = imputer.transform(X[:, 1:3])

print(X)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# 将变量编码
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])

onehotencoder = OneHotEncoder(categorical_features = [0])
X = onehotencoder.fit_transform(X).toarray()

labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
# 拆分训练集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
# 特征缩放
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```
